## Is there A Correlation between Age and purchase Amount?

```python
data_frame["Age"].corr(data_frame["Purchase Amount (USD)"])
```

$$H_0: r = 0$$
$$H_1: r \neq 0$$
$$\alpha = 0.01$$

```python
alpha = 0.01
```

```python
n=data_frame.shape[0]
r = np.corrcoef(data_frame["Age"], data_frame["Purchase Amount (USD)"])[0, 1]

t_value = r / np.sqrt((1 - r**2) / (n - 2))
t_value
```

```python
from scipy.stats import t

degrees_freedom = n - 2

p_value_lower = t.cdf(-np.abs(t_value), degrees_freedom)
p_value_upper = 1 - t.cdf(t_value, degrees_freedom)
p_value = p_value_lower + p_value_upper
p_value
```

```python
p_value < alpha
```

### Since The P_value > $\alpha$, we fail to reject the null hypothesis and conclude that there is **no** **correlation** between Age and Purchase Amount in USD

### No Correlation Between Age and Purchase Amount

## K-Mediods

```python
from sklearn_extra.cluster import KMedoids

X = data_frame[['Age', 'Purchase Amount (USD)']]

k_values = range(1, 6)  
wcss = [] 

for k in k_values:
    kmedoids = KMedoids(n_clusters=k, metric='manhattan', random_state=42)
    kmedoids.fit(X)
    wcss.append(kmedoids.inertia_)  

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_values, wcss, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method for K-Medoids Clustering')
plt.grid(True)
plt.show()
```

### The Elbow Method Indicates that using 3 clusters will result in the best outcome

```python
X = data_frame[['Age', 'Purchase Amount (USD)']]

k=3
kmedoids = KMedoids(n_clusters=k).fit(X)

X.insert(0, 'kmedoids Cluster Labels', kmedoids.labels_)
X.head()
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(X, x="Age", y='Purchase Amount (USD)', hue="kmedoids Cluster Labels", style="kmedoids Cluster Labels", palette="deep")
```
